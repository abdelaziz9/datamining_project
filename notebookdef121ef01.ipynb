```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

Load the data

```python
train = pd.read_csv('/kaggle/input/titanic/train.csv', na_values='NAN')
train.head(10)
ageMean = round(train.Age.mean())
train.Age.fillna(ageMean , inplace= True)
```

```python
test = pd.read_csv('/kaggle/input/titanic/test.csv')
test.head()
ageMeantest = round(test.Age.mean())
test.Age.fillna(ageMeantest , inplace= True)
test.Age.head(15)
```

Import the model we are going to use which is RandomForestRegressor , and the model evaluation which is mae


```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

extraxt survived values and put it in Y


```python
y = train.Survived
y.head()
```

Extract features and put it in variable x
Also , the pd_getdummies function is used to change sex column values from (male, female) into (1,0)

```python
features = ["Pclass","Sex","SibSp","Parch"]
x= pd.get_dummies(train[features])
x.head(10)
```

Extract features from test data sett

```python
testx = pd.get_dummies(test[features])
testx.head(10)
```

splitting the data into 80% training set and 20% validation set

```python
trainx, valx, trainy, valy = train_test_split(x, y, random_state = 0)
```

Building model using randomforestregressor with maximum depth of 5

```python

model = RandomForestRegressor(max_depth = 5, random_state=1)

model.fit(trainx,trainy)

```

the model we are using is mean absolute error , we get an error rate by about 27%

```python
predictionvalidation = model.predict(valx)
print ( "mean absolute error is" , mean_absolute_error(valy,predictionvalidation ))
```

then roudning the values and changing them into integer, i tried to submit wihout rounding values and converting them into integer i got 0% score, so i had to do this step

```python
prediction = model.predict(testx)
integerprediction = prediction.round().astype(int)
integerprediction
output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': integerprediction})
output.to_csv('my_submission.csv', index=False)
```
